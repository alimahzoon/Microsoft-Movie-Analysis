```python
#this is the main function that we can use
def func_genre_list(x):
    number_of_iterate = range(9998)
    genres_in_row = []
    for n in number_of_iterate:
        test = ast.literal_eval(x[n])
        simple_list = []
        for i in range(len(test)):
            simple_list.append(test[i]['name'])
        genres_in_row.append(simple_list)
    return genres_in_row
```

```python
list_of_genres = func_genre_list(movie_df['genres'])
```
