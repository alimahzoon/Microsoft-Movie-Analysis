```python
#this is the main function that we can use
#Lets get the genres out of their dictionaries
def func_genre_list(x):
    number_of_iterate = range(9998)
    genres_in_row = []
    for n in number_of_iterate:
        test = ast.literal_eval(x[n])
        simple_list = []
        for i in range(len(test)):
            simple_list.append(test[i]['name'])
        genres_in_row.append(simple_list)
    return genres_in_row
```

```python
#Apply the function
list_of_genres = func_genre_list(movie_df['genres'])
```

```python
#create a new column 
#There is nothing to be worry about this error...
movie_df['genre_list'] = list_of_genres
```

```python
movie_df.head()
```

```python
#Create a new DF that only contains 2 columns.
title_genre = movie_df.loc[:,['title', 'genre_list']]
```

```python
#To remove brackets 
title_genre['genre_list'] = title_genre.genre_list.map(lambda x:", ".join( repr(e) for e in x ) )
```

```python
title_genre.head()
```

```python
#To remove the Quotation mark
title_genre['genre_list'] =title_genre.genre_list.map(lambda x: x.replace("'", '') )
```

```python
#To remove spaces between strings
title_genre['genre_list'] =title_genre.genre_list.map(lambda x: x.replace(" ", '') )
```

```python
#Each movie has a combnition of genres(up to 6 genres!); lets Split them up
col_sep_df = title_genre['genre_list'].str.get_dummies(sep=',')
```

```python
#Create a new DF with each genre included
new_df =pd.concat([title_genre,col_sep_df], axis=1, join='outer')
```
